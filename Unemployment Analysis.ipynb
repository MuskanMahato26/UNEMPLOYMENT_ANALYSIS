```python
#Estimated Labour Participation Rate:
#The Labour Participation Rate (LPR) is the percentage of the working-age population (people who are either employed or actively seeking employment) that is either employed or actively looking for work.
#It is calculated as the sum of employed individuals and those actively seeking employment divided by the total working-age population, multiplied by 100.
#A higher labour participation rate generally indicates a larger proportion of the population actively involved in the labor market.

 #Estimated Employed:
#This refers to the number of people who are currently employed in the workforce.
#It includes individuals who have jobs, regardless of whether they are working full-time, part-time, or in temporary positions.

 #Estimated Unemployment Rate:
#The Unemployment Rate is the percentage of the labor force that is unemployed and actively seeking employment.
#It is calculated as the number of unemployed individuals divided by the total labor force (the sum of employed and unemployed individuals), multiplied by 100.
#A higher unemployment rate indicates a higher proportion of the labor force without employment.
```

```python
#see the unemployment rate according to different regions of india.
```

```python
data_cleaned.columns =["States", "Date", "Frequency", "Estimated Unemployment Rate",
                       "Estimated Employed","Estimated Labour Participation Rate",
                       "Region"]
plt.figure(figsize=(10,8))
plt.title("India's Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data =data_cleaned)
plt.show()
```

```python


# Matplotlib subplot
fig = plt.figure()
axe = fig.add_subplot(1, 2, 1)

# Top 10 States with highest unemployment using Matplotlib
data_cleaned[:10].plot(kind='bar', color='red', figsize=(30, 5), ax=axe)

axe.set_title('Top 10 States with highest unemployment')
axe.set_xlabel('States')
axe.set_ylabel('Estimated Unemployment Rate')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Bar plot for Labor Participation Rate
axes[0].bar(data_cleaned["Date"], data_cleaned["Estimated Labour Participation Rate"], color='blue', label='Labor Participation Rate')
axes[0].set_title('Labor Participation Rate Each Month')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Estimated Labor Participation Rate (%)')
axes[0].legend()

# Bar plot for Unemployment Rate
axes[1].bar(data_cleaned["Date"], data_cleaned["Estimated Unemployment Rate"], color='red', label='Unemployment Rate')
axes[1].set_title('Unemployment Rate Each Month')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Estimated Unemployment Rate')
axes[1].legend()

plt.tight_layout()
plt.show()
```
